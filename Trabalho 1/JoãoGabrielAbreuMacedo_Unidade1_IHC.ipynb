### Playstation 5
#### Coisas que gosto:
 - Poder jogar na televisão
 - Jogar com minha esposa

#### Coisas que não gosto
 - Baixa quantidade de jogos novos
 - Preços de jogos e do serviço online é muito alto.


### Ar condicionado portátil
Coisas que gosto:
 - Não precisa fazer obra para poder utilizar
 - esfria bem o ambiente

Coisas que não gosto
 - Gasta muito mais energia que um AC normal
 - Faz muito barulho
 - Ocupa muito espaço
 - precisa descarregar a água regularmente


### Aspirador robô
Coisas que gosto
 - Ajuda na limpeza
 - Dá uma folga à minha esposa, ela fica feliz

Coisas que não gosto
 - Precisa preparar bem o ambiente
 - não passa bem em tapetes
 - não limpa muito bem, mesmo no modo de passar pano


### TV
Coisas que gosto
 - Tamanho é bom, nem muito grande nem muito pequena
 - é smart, posso assistir serviços de streaming

 Coisas que não gosto
 - UI muito confusa
 - processador lentíssimo
 - processo de conectar no wi-fi é muito lento e complicado
 - alexa no controle não funciona

### Carro à combustão
Coisas que gosto
 - rápido
 - me leva pra onde preciso ir, grande autonomia

Coisas que não gosto
 - 300 reais para encher o tanque
 - Multimídia quebrou
 - Baixo, pega em qualquer meio-fio de estacionamento
